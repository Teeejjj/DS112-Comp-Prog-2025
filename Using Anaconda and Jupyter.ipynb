# Jupyter and Anaconda

**What is Jupyter?** <br>
Jupyter Notebook (formerly known as IPython Notebook) is an interactive web application for creating and sharing computational documents. The project was first named IPython and later renamed Jupyter in 2014. It is a fully open-source product, and users can use every functionality available for free. [üîó](https://domino.ai/data-science-dictionary/jupyter-notebook)

**Its Data Scientists Best Friend** <br>
Jupyter is an interactive tool that lets you write, run, and share your code and notes together in a single document, making it perfect for learning, analyzing data, and telling stories. <br>
Jupyter is popular because it supports the exploratory and iterative workflow of data science perfectly. For a new student, it lowers the barrier to entry, making it easier to experiment, learn from mistakes, and clearly communicate your findings‚Äîall in one place.

**Whats the difference between Jupyter Notebook and JupyterLab?** <br>
Both `JupyterLab` and `Jupyter Notebook` let you use languages like Python, Julia, Scala, and R. The difference is that Jupyter Notebook has a very simple setup‚Äîyou can open notebooks, text files, and terminals. JupyterLab, however, gives you a lot more tools in one place. You can use consoles, edit CSV and markdown files, work with interactive maps, and even drag and drop notebook cells. Think of JupyterLab as an upgraded version of Jupyter Notebook that makes coding and exploring data easier and more flexible.

**What is Anaconda?** <br>
Anaconda is a free, open-source distribution (or ‚Äúpackage‚Äù) that makes it easy to work with Python and R for data science, programming, and machine learning.
<br><br>
Think of it as a toolbox that already comes with:
- Python (The progamming language)
- Useful Libraries (like Numpy, Pandas, Matplotlib)
- Jupyter Notebook and JupyterLab
- package manager `conda` (install/update packages)

**Whats a Variable?** <br>
A **variable** is like a container where you can store information (data) that your program can use later.

A variable can have a short name (like x and y) or a more descriptive name (age, car_name, 
total_volume). 
- Rules for Python variables:
 - A variable name must start with a letter or the underscore character
 - A variable name cannot start with a number
 - A variable name can only contain alphanumeric characters and underscores (A-z, 0-9, and _)
 - Variable names are case-sensitive (age, Age and AGE are three different variables)
 
 > ****NOTE****: <u>variable names are case-sensitive</u>

```python
x = 18 # We assigned the number 25 to `x`
y = "My age is" # We also assigned the string value "My age is" to y
print(y)
print(x)
```

**String? Integer?** <br>
Data can be numerical or textual, and that knowing which is which is important. <br>
A `data type` tells Python (or any programming language) what kind of value a variable is holding.

| Dtype |     |
|------ | ----|
| int or integer | `Whole Numbers` - 1 , 2 , 10 , 26 , 100 , 500 , 100000 |
| floating point or float | `Numbers with Decimals` - 1.1, 1.2, 27.8, 100.10|
| String or str | `Text` - "Hello", "World", "Hello World!!" |
| Boolean or bool | `True or False`|
| Lists | `Ordered collection of items` - ['apple', 'oranges', 'grapes'] |
| Tuples | `Ordered but unchangeable collection` - (10,20,30) |
| Dictionary | `Key-value pairs` - {'Name1': 'Alvin', 'Name2': 'Simon', 'Name3':'Theodore'} |

```python
x = "Wednesday"
y = 18
print(x,"age is",y)
```

```python
fruits = ['oranges', 'apples', 'watermelon', 'grapes']
for x in fruits:
    print(x)
```

```python
names = {'Name1': 'Alvin', 'Name2': 'Simon', 'Name3':'Theodore'}
for k,v in names.items():
    print(k + " " + v)
```
